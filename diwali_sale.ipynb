```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Amount"]=df["Amount"].fillna(df["Amount"].mean())
```

```python
df.isnull().sum()
```

```python
#change data type
df["Amount"]=df["Amount"].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python

# ax.containers kya hota hai?

# Answer:
# ax.containers Matplotlib ke andar maujood sabhi BarContainer objects ki list return karta hai. 
#Har BarContainer ek group of bars ko represent karta hai.

# ax.bar_label() kya karta hai?

# Answer:
# ax.bar_label() har bar ki height (value) ko automatically read karke uske upar label display karta hai.
#Isse manually text() ya annotate() likhne ki zarurat nahi padti.

# Ye bar_label() method Matplotlib 3.4+ me introduce hua tha aur bar chart par 
#values dikhane ka sabse simple aur recommended tareeka hai.
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen= df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
ax= sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
#Tatal Amount Vs Age Group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y = 'Amount',data=sales_age)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y = 'Amount',
           data=sales_age,
           hue='Age Group',
           palette='Set2',
           legend=False)
plt.show()
```

```python
# "Analysis se yeh observe hota hai ki 26–35 years age group ke customers sabse zyada purchases karte hain,
# aur is age group me female buyers ka contribution sabse adhik hai."
```

```python
#State
#Total number of orders from top 10 States
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```
